```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("cars.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

#### Que #1: Find the population mean and standard deviation for peakrpm column.

```python
pop_mean = df['peakrpm'].mean()
pop_std = df['peakrpm'].std()

print("The population mean is", pop_mean)
print("The population standard deviation is", pop_std)
```

#### Que #2: Draw a sample of 30 from the population of peakrpm column and calculate the sample mean and standard deviation.

```python
sample = df.sample(n=30, random_state=1)
```

```python
sample_mean = sample['peakrpm'].mean()
sample_std = sample['peakrpm'].std()

print("The sample mean is", sample_mean)
print("The sample standard deviation is", sample_std)
```

#### Que #3: Calculate the margin of error for 95% confidence level for peakrpm using standard deviation of sample.

```python
n = 30
z_critical = stats.norm.ppf(q=0.95)

margin_of_error = z_critical * (sample_std / n**0.5)
print("The margin of error is", margin_of_error)
```

#### Que #4: Calculate confidence interval for the peakrpm.

```python
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print("The confidence interval is", confidence_interval)
```
