# <center>Paired Sample T - Test</center>

#### Example I

We have a dataset which is called 'love breakups' csv file and it contains the alcohol consumption between and after breakup.   
Now conduct a paired sample t-test to check whether the alcohol consumption is more after the breakup at alpha equal to 0.05.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("love_breakups.csv")
df.head()
```

```python
sns.distplot(df['pre_Breakup'], label="Before Breakup")
sns.distplot(df['post_Breakup'], label="After Breakup")

plt.legend()
plt.show()
```

```python
stats.ttest_rel(df['pre_Breakup'], df['post_Breakup'])
```

Since, the p-vale is greater than 0.05, Null Hypothesis follows.  
Therefore, the alcohol consumption is more after the breakup.
