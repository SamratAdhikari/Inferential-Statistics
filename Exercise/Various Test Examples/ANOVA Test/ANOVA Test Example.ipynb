# <center>ANOVA Test</center>

#### Example I

We have a brand manager at Dishy which is a detergent company and he is going to offer the discounts.  
The discounts are 0%, 10% and 20% on randomly selected days.  
Now, we have the data that contains the quantity sold in kilograms of the detergent powder on each day.  
As part of the hypothesis testing, we want to perform one way ANOVA to check whether the discount has any impact on the quantity of detergent sold.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("dishy.csv")
df.head()
```

```python
sns.distplot(df['discount_0'], label="Discount: 0")
sns.distplot(df['discount_10'], label="Discount: 10")
sns.distplot(df['discount_20'], label="Discount: 20")

plt.legend()
plt.show()
```

```python
stats.f_oneway(df['discount_0'], df['discount_10'], df['discount_20'])
```

```python
round(6.3178284112882e-18, 4)
```

The p-value is less than 0.05.  
Discount has a statistically significant impact on the quantity of items sold.
